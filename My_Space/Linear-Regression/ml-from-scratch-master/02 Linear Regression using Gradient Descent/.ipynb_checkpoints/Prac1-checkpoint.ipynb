```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
X =data.iloc[:,0]
Y = data.iloc[:,1]
plt.scatter(X,Y)
```

```python
m = 0
c = 0
L = 0.0001
epoch = 1000
n = len(X)

for i in range(n):
    Y_pred = m*X + c
    D_m = (-1/n)*sum(X*(Y-Y_pred))
    D_c = (-1/n)*sum(Y-Y_pred)
    m = m-L*D_m
    c = c-L*D_c
print(m,c)
```
