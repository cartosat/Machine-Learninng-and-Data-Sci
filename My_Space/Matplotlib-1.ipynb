```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(1,10,10)
y = x**2
z = x**3
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-')
plt.xlabel("X--Lable")
plt.ylabel("Y--Lable")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

#plt.subplot(1,2,2)
#plt.plot(y,x,'b')

plt.subplot(1,2,2)
plt.plot(x,z,'b')

plt.show()
```

```python
#Object Oriented Method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
plt.show()
```

```python
a = np.linspace(1,5,8)
b = np.log(a)
```

```python
plot=plt.figure()
graph=plot.add_axes([0.1,0.1,0.8,0.8])
graph.plot(a,b)
```

```python
plot = plt.figure()

graph1 = plot.add_axes([0.1,0.1,0.8,0.8])
graph2 = plot.add_axes([0.5,0.1,0.4,0.3])

graph1.plot(a,b)
graph2.plot(b,a)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First')

axes[1].plot(y,x,'r')
axes[1].set_title('Second')
plt.tight_layout()
plt.show()
```

```python
data = plt.figure()
graph = data.add_axes([0,0,1,1])
graph.plot(x,y,label='X vs Y')
graph.plot(y,x,label='Y vs X')
graph.legend()
```
