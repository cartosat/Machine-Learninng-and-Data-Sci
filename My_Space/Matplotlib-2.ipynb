```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(1,10,10)
y = x**2
```

```python
fig = plt.figure()
var = fig.add_axes([0,0,1,1])
var.plot(x,y,lw=2,color='purple',ls='--',marker='o',markerfacecolor='red',markersize=10)
```
