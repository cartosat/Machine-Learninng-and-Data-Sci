```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue="sex",markers=['o','v'],aspect=0.8,height=8)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.set_contex('poster')
sns.countplot(x='sex',data=tips)
```
