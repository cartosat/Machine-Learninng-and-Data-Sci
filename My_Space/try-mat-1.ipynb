```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,20,10)
y = np.log(x)
```

```python
#inbuild Function method
```

```python
plt.plot(x,y)
plt.show()
```

```python
#Object oriented method
```

```python
plot = plt.figure()
graph=plot.add_axes([0.1,0.1,0.8,0.8])
graph.plot(x,y,'r')
```

```python
#subplot
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'green')

plt.subplot(1,2,2)
plt.plot(y,x,'purple')
```

```python
#OOP Method for sub-plot
```

```python
fig = plt.figure()

graph1 = fig.add_axes([0.1,0.1,0.8,0.8])
graph2 = fig.add_axes([0.5,0.1,0.4,0.4])

graph1.plot(x,y,'r',lw=0.9)
graph1.set_title('Larger Graph',color='green')

graph2.plot(y,x)
graph2.set_title('Smaller Graph',color='green')
```

```python
dyan = plt.figure()
figure1 = dyan.add_axes([0.1,0.1,0.8,0.8])
figure2 = dyan.add_axes([0.5,0.1,0.4,0.4])
figure1.plot(x,y)
figure1.set_title('Larger',color='green')
figure2.plot(x,y,'r')
figure2.set_title('Smaller',color='green')
plt.show()
```
